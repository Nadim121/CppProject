```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import BernoulliNB
```

```python
df = pd.read_csv('classification.csv')
df.head(8)
```

```python
print(df.groupby('Category').describe())
```

```python
df['spam'] = df['Category'].apply(lambda x: 1 if x=='spam' else 0)
```

```python
# prompt: remove the "Category" column from df data matrix

df = df.drop('Category', axis=1)
```

```python

df
```

```python
# prompt: create train/test split

X_train, X_test, y_train, y_test = train_test_split(
    df['Message'], df['spam'], test_size=0.25, random_state=42)
```

```python
X_train.describe()
```

```python
cv = CountVectorizer()
X_train_count = cv.fit_transform(X_train.values)
X_train_count.toarray()[:5]
```

```python
model = MultinomialNB()
model.fit(X_train_count, y_train)
```

```python
model = BernoulliNB()
model.fit(X_train_count, y_train)
```

```python
X_test_count = cv.transform(X_test)
print("Accuracy = ",model.score(X_test_count, y_test)*100 , "%")
```
