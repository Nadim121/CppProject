```python
with open('SMSSpamCollection.txt', 'r') as f:
    data = f.read()

data[:300]
```

```python
# Remove punctuation and lowercase
from string import punctuation
all_text = ''.join([c for c in data if c not in punctuation])
all_text = all_text.lower()

# split label and text of each line.
messages = all_text.split('\n')
messages = [x.split('\t') for x in messages if len(x)>=1]
[labels, texts] = np.array([list(x) for x in zip(*messages)])
```

```python
print("Example: ")
print("Label: {},\tText: {}".format(labels[0],texts[0]))
print("Label: {},\tText: {}".format(labels[1],texts[1]))
print("Label: {},\tText: {}".format(labels[2],texts[2]))
```

```python
labels = np.array([1 if each == 'spam' else 0 for each in labels])
labels
```

```python
# All words
all_text = ' '.join(texts)
words = all_text.split()
words[:20]
```

```python
from collections import Counter
counts = Counter(words)
vocab = sorted(counts, key=counts.get, reverse = True)
vocab_to_int = {word: ii for ii, word in enumerate(vocab)}
```

```python
vocab_to_int
```

```python
# Convert the reviews to integers, same shape as reviews list, but with integers
text_ints = []
for each in texts:
    text_ints.append([vocab_to_int[word] for word in each.split()])
```

```python
text_ints[3]
```

```python
from collections import Counter
text_lens = Counter([len(x) for x in text_ints])
print("Zero-length text: {}".format(text_lens[0]))
print("Maximum text length: {}".format(max(text_lens)))
```

```python
non_zero_idx = [ii for ii, texts in enumerate(text_ints) if len(texts) != 0]
len(non_zero_idx)
```

```python
len(texts)
```

```python
# fillter out that review with 0 length
text_ints = [text_ints[ii] for  ii in non_zero_idx]
labels = np.array([labels[ii] for ii in non_zero_idx])
```

```python

seq_len = 170
features = np.zeros((len(text_ints), seq_len), dtype=int)
for i, row in enumerate(text_ints):
    features[i, -len(row):] = np.array(row)[:seq_len]
```

```python
features[0]
```

```python
split_frac = 0.8
split_idx = int(len(features)*0.8)
train_x, val_x = features[:split_idx], features[split_idx:]
train_y, val_y = labels[:split_idx], labels[split_idx:]

test_idx = int(len(val_x)*0.5)
val_x, test_x = val_x[:test_idx], val_x[test_idx:]
val_y, test_y = val_y[:test_idx], val_y[test_idx:]

print("\t\t\tFeature Shapes:")
print("Train set: \t\t{}".format(train_x.shape),
      "\nValidation set: \t{}".format(val_x.shape),
      "\nTest set: \t\t{}".format(test_x.shape))
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from sklearn.model_selection import train_test_split
import numpy as np

# Assuming `features` and `labels` are prepared from the earlier preprocessing
seq_len = 170
vocab_size = len(vocab_to_int) + 1  # Vocabulary size (+1 for padding token)
embedding_dim = 128
lstm_units = 128
dropout_rate = 0.5
batch_size = 64
epochs = 10

# Split the data into training and testing sets
train_x, test_x, train_y, test_y = train_test_split(features, labels, test_size=0.2, random_state=42)

# Build the LSTM Model
model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=seq_len),
    LSTM(lstm_units, return_sequences=False, dropout=dropout_rate, recurrent_dropout=dropout_rate),
    Dense(64, activation='relu'),
    Dropout(dropout_rate),
    Dense(1, activation='sigmoid')  # Binary classification
])

# Compile the Model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Model Summary
model.summary()

# Train the Model
history = model.fit(train_x, train_y, validation_data=(test_x, test_y), batch_size=batch_size, epochs=epochs, verbose=1)

# Evaluate the Model
loss, accuracy = model.evaluate(test_x, test_y, verbose=1)
print(f"Test Accuracy: {accuracy:.2f}")

# Save the Model
model.save("lstm_spam_classifier.h5")

# Load and Test the Model on New Messages
loaded_model = tf.keras.models.load_model("lstm_spam_classifier.h5")

# Example Messages to Test
new_messages = ["Congratulations! You've won a free vacation. Call now!", "Can we schedule the meeting for next week?"]
new_sequences = [[vocab_to_int[word] if word in vocab_to_int else 0 for word in message.split()] for message in new_messages]
new_features = np.zeros((len(new_sequences), seq_len), dtype=int)
for i, row in enumerate(new_sequences):
    new_features[i, -len(row):] = np.array(row)[:seq_len]

# Predictions
predictions = loaded_model.predict(new_features)
for msg, pred in zip(new_messages, predictions):
    label = "Spam" if pred > 0.5 else "Ham"
    print(f"Message: \"{msg}\" - Prediction: {label}")
```
